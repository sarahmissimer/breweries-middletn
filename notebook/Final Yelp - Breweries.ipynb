```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from bs4 import BeautifulSoup as BS

%matplotlib inline
```

```python
url = "https://api.yelp.com/v3/businesses/search"
key = open(r'C:\Users\sarah\Documents\DA7\projects\repos\capstone\keys\Yelp API Key.txt').readlines()[0]
headers = {
    'Authorization': 'Bearer %s' % key
}
```

```python
parameters = {'location': 'Nashville, TN',
             'term': 'Beer Brewery Brewers Brewer',
             'radius': 10000,
             'limit':50,
            }

response = requests.get(url, headers=headers, params=parameters)
response
```

```python
response.json()
```

```python
yelp_info = pd.DataFrame(response.json()['businesses'])
yelp_info

yelp_info_df = yelp_info['location'].apply(pd.Series)
yelp_info_df

yelp_info_df1 = yelp_info_df[['address1','city','zip_code','state']]
yelp_info_df1.head()
```

```python
yelp_infoa = yelp_info[['name','rating','review_count']]
yelp_infoa.head()
```

```python
breweries_yelp = pd.concat([yelp_infoa, yelp_info_df1], axis=1, join='inner')
breweries_yelp
```
