```python
zipcodes_df = pd.read_csv("../data/zipcodes_df.csv")
zipcodes_df.head()
```

```python
zipcodes = zipcodes_df[['zip_code','2011 total household','2011 household median income', '2011 Median Value', '2011 population',
                       '2015 total household', '2015 household median income', '2015 Median Value', '2015 population',
                       '2020 total household', '2020 household median income', '2020 Median Value', '2020 population']]

zipcodes = zipcodes.replace(',', '', regex=True)
zipcodes

zipcodes = zipcodes.replace('-', '0', regex=True)
zipcodes

zipcodes['zip_code'] = zipcodes[['zip_code']].astype(str)
zipcodes

zipcodes[['2011 total household','2011 household median income', '2011 Median Value', '2011 population','2015 total household', '2015 household median income', '2015 Median Value', '2015 population','2020 total household', '2020 household median income', '2020 Median Value', '2020 population']] = zipcodes[['2011 total household','2011 household median income', '2011 Median Value', '2011 population', '2015 total household', '2015 household median income', '2015 Median Value', '2015 population','2020 total household', '2020 household median income', '2020 Median Value', '2020 population']].astype(int)
zipcodes

zipcodes = zipcodes.drop(zipcodes.index[36]).drop(zipcodes.index[37])
zipcodes
```
