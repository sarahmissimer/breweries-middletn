```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from functools import reduce

%matplotlib inline
pd.options.display.max_columns = None
# pd.options.display.max_rows = None
```

```python
housing_value_df = pd.read_csv("../data/value_dataframe.csv")
```

```python
population_df = pd.read_csv("../data/population_dataframe.csv")
```

```python
income_df = pd.read_csv("../data/income_dataframe.csv")
```

```python
zipcodes_df = reduce(lambda left, right:
                      pd.merge(left, right,
                              on=['zip_code']),
                      [income_df, population_df, housing_value_df])
zipcodes_df.head()
```

```python
zipcodes_df_final = zipcodes_df.drop([15, 19, 20, 16, 21, 18, 22, 28, 25, 34, 36, 37, 35])
zipcodes_df_final.info()
```

```python

zipcodes_df_final
```

```python
zipcodes_df_final = zipcodes_df_final.replace(',', '', regex=True)
zipcodes_df_final
```

```python
growth_df = zipcodes_df_final[['zip_code', '2011 population', '2015 population', '2020 population']]
growth_df.info()
```

```python
growth_df[['2011 population','2015 population','2020 population']] = growth_df[['2011 population','2015 population','2020 population']].astype(int)
growth_df
```

```python
growth_df[['zip_code']] = growth_df[['zip_code']].astype(str)
```

```python
growth_df['average growth'] = growth_df[['2011 population','2015 population','2020 population']].mean(1).round()

growth_df['growth rate'] = (growth_df['2020 population']-growth_df['2011 population'])/growth_df['2011 population']*100

growth_df
```

```python
breweries_df = pd.read_csv("../data/breweries_final_dataframe.csv")
breweries_df.head()
```

```python
breweries_df = breweries_df.replace(',', '', regex=True)
breweries_df
```

```python
breweries_df['zip_code'].unique()
```

```python
breweries_df[['zip_code']] = breweries_df[['zip_code']].astype(str)
```

```python
breweries_growth_df = breweries_df[['zip_code', '2011 population', '2015 population', '2020 population']]
```

```python
breweries_growth_df[['2011 population','2015 population','2020 population']] = breweries_growth_df[['2011 population','2015 population','2020 population']].astype(int)
breweries_growth_df
```

```python
breweries_growth_df['average growth'] = breweries_growth_df[['2011 population','2015 population','2020 population']].mean(1).round()

breweries_growth_df['growth rate'] = (breweries_growth_df['2020 population']-breweries_growth_df['2011 population'])/breweries_growth_df['2011 population']*100

breweries_growth_df
```

```python
bgdf= breweries_growth_df.groupby('zip_code', as_index=False)['2011 population', '2015 population', '2020 population', 'average growth', 'growth rate'].last()
bgdf

bgdf = bgdf.drop(bgdf.index[7])
bgdf
```

```python
x = bgdf['zip_code']
y = bgdf['average growth']

fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(x, y, color ='maroon',
        width = 0.4)
 
plt.xlabel("Zip Codes")
plt.ylabel("Average Growth")
plt.title("Growth of Zip Codes with Breweries")
plt.show();
```

```python
x = growth_df['zip_code']
y = growth_df['average growth']

fig = plt.figure(figsize = (20, 5))
 
# creating the bar plot
plt.bar(x, y, color ='maroon',
        width = 0.4)
 
plt.xlabel("Zip Codes")
plt.ylabel("Average Growth")
plt.title("Growth of Zip Codes without Breweries")
plt.show();
```
