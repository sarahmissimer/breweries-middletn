```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from functools import reduce
import math

%matplotlib inline
# pd.options.display.max_columns = None
# pd.options.display.max_rows = None
```

```python
building_permits = pd.read_csv("../data/Building_Permits_Issued.csv", low_memory=False)
building_permits
```

```python
building_permits.info()
```

```python
#permit type description, date entered, date issued, address, city, state, ZIP

building_permits = building_permits[['Permit Type Description','Date Entered','City', 'ZIP']]
building_permits
```

```python
building_permits['ZIP'] = building_permits['ZIP'].astype(str)
```

```python
building_permits.info()
```

```python
building_permits1 = building_permits.rename(columns={'Permit Type Description':'Permit Type', 'ZIP':'zip_code'})
building_permits1
```

```python
bp1 = building_permits1.loc[(building_permits1['City'] == "NASHVILLE")]
bp1
```

```python
bp1['Date Entered'] = pd.to_datetime(bp1['Date Entered'])

bp1.info()
```

```python
bp2 = bp1.loc[(bp1['Date Entered'] >= '2017-01-01')]
bp2
```

```python
oldest = min(bp1['Date Entered'])
oldest
```

```python
zip_codes1 = []
zc = bp1['zip_code']

for z in zc:
    zip_codes1.append(str(z).split('.')[0])
    
final_zips = [int(z) for z in zip_codes1]
final_zips
```

```python
df1 = pd.DataFrame(final_zips, columns = ['zip_code'])
df1
```

```python
df2 = bp1[['Permit Type','Date Entered']]
df2
```

```python
permits_df = pd.concat([df1, df2], axis=1, join='inner')
permits_df
```

```python
len(permits_df['zip_code'].unique())
```

```python
permits_df.groupby('zip_code', as_index=False).last()
```

```python
permits_df_gb = permits_df.groupby('zip_code')['zip_code'].count()
permits_df_gb
```
